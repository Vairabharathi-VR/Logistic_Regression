# Importing the basic necessary packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

# Extracting the data 

```python
def read_data():
    path = 'E:/Logistic Regresssion_Project7/bank-full.csv'
    data = pd.read_csv(path, sep=';')
    return data
```

```python
data = read_data()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
for column in data.select_dtypes(include=['object']).columns:
    data[column] = label.fit_transform(data[column])
```

```python
data
```

# Model selection

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop('y', axis=1)
y = data['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
model = LogisticRegression()
model_fit = model.fit(X_train, y_train)
pred = model_fit.predict(X_test)
acc = accuracy_score(y_test, pred)
print(f'The accuracy of the model is {acc}')
```

```python
# Model accuracy is so low what we can do now is Scaling and feature selection
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y ,random_state=465)
```

```python
model = LogisticRegression()
model_fit = model.fit(X_train, y_train)
pred1 = model_fit.predict(X_test)
acc = accuracy_score(y_test, pred1)
print(f'The Accuracy of the model is {acc}')
```
